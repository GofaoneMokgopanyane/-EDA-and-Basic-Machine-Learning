```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from scipy.stats import chi2_contingency
import warnings
%matplotlib inline
warnings.filterwarnings('ignore')

# Set style for better visualizations
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
```

```python


# Paths
processed_path = '../data/processed/cleaned_seismic_data.csv'
raw_path = '../data/raw/seismic-bumps.csv'
eda_dir = '../data/eda'
os.makedirs(eda_dir, exist_ok=True)

# Load both raw and processed data
df_raw = pd.read_csv(raw_path)
df_processed = pd.read_csv(processed_path)

print("=" * 80)
print("SEISMIC BUMPS DATASET - EXPLORATORY DATA ANALYSIS")
print("=" * 80)

# 1. BASIC DATASET OVERVIEW
print("\n" + "="*40)
print("1. DATASET OVERVIEW")
print("="*40)
print(f"Dataset shape: {df_raw.shape}")
print(f"\nFirst 5 rows:")
print(df_raw.head())
print(f"\nData types:")
print(df_raw.dtypes)
print(f"\nBasic statistics:")
print(df_raw.describe())

# 2. CLASS DISTRIBUTION ANALYSIS
print("\n" + "="*40)
print("2. TARGET VARIABLE ANALYSIS (Hazardous vs Safe)")
print("="*40)

class_counts = df_raw['class'].value_counts()
print(f"\nClass distribution:")
print(f"  Class 0 (Safe): {class_counts[0]} ({class_counts[0]/len(df_raw)*100:.2f}%)")
print(f"  Class 1 (Hazardous): {class_counts[1]} ({class_counts[1]/len(df_raw)*100:.2f}%)")
print(f"  Imbalance ratio: {class_counts[0]/class_counts[1]:.2f}:1")
```

```python

# Plot class distribution
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Bar plot
axes[0].bar(['Safe (0)', 'Hazardous (1)'], class_counts.values, 
            color=['#2ecc71', '#e74c3c'])
axes[0].set_title('Class Distribution', fontsize=14, fontweight='bold')
axes[0].set_ylabel('Count')
for i, v in enumerate(class_counts.values):
    axes[0].text(i, v + 50, f'{v}\n({v/len(df_raw)*100:.1f}%)', 
                 ha='center', fontweight='bold')

# Pie chart
axes[1].pie(class_counts.values, labels=['Safe (0)', 'Hazardous (1)'], 
            autopct='%1.2f%%', colors=['#2ecc71', '#e74c3c'], startangle=90)
axes[1].set_title('Class Distribution (%)', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.savefig(os.path.join(eda_dir, 'class_distribution.png'), dpi=300, bbox_inches='tight')
plt.show()
```

```python

# 3. CATEGORICAL VARIABLES ANALYSIS
print("\n" + "="*40)
print("3. CATEGORICAL VARIABLES ANALYSIS")
print("="*40)

categorical_cols = ['seismic', 'seismoacoustic', 'shift', 'hazard']

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.ravel()

for idx, col in enumerate(categorical_cols):
    # Count plot
    counts = df_raw[col].value_counts()
    axes[idx].bar(counts.index, counts.values, alpha=0.7)
    axes[idx].set_title(f'Distribution of {col}', fontsize=12, fontweight='bold')
    axes[idx].set_xlabel(col)
    axes[idx].set_ylabel('Count')
    
    # Add value labels
    for i, v in enumerate(counts.values):
        axes[idx].text(i, v + max(counts.values)*0.01, str(v), ha='center')
    
    # Print distribution
    print(f"\n{col} distribution:")
    for val, count in counts.items():
        print(f"  {val}: {count} ({count/len(df_raw)*100:.1f}%)")

plt.tight_layout()
plt.savefig(os.path.join(eda_dir, 'categorical_distributions.png'), dpi=300, bbox_inches='tight')
plt.show()
```

```python

# 4. CATEGORICAL VARIABLES vs TARGET
print("\n" + "="*40)
print("4. CATEGORICAL FEATURES vs TARGET (Hazard Analysis)")
print("="*40)

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.ravel()

for idx, col in enumerate(categorical_cols):
    # Cross-tabulation
    crosstab = pd.crosstab(df_raw[col], df_raw['class'], normalize='index') * 100
    
    # Plot
    crosstab.plot(kind='bar', ax=axes[idx], color=['#2ecc71', '#e74c3c'], alpha=0.7)
    axes[idx].set_title(f'Hazard Rate by {col}', fontsize=12, fontweight='bold')
    axes[idx].set_xlabel(col)
    axes[idx].set_ylabel('Percentage (%)')
    axes[idx].legend(['Safe', 'Hazardous'])
    axes[idx].axhline(y=crosstab[1].mean(), color='red', linestyle='--', alpha=0.5, label='Avg Hazard Rate')
    
    # Print statistics
    print(f"\nHazard rate by {col}:")
    hazard_rates = df_raw.groupby(col)['class'].mean() * 100
    for val, rate in hazard_rates.items():
        count = df_raw[df_raw[col] == val].shape[0]
        hazardous_count = df_raw[(df_raw[col] == val) & (df_raw['class'] == 1)].shape[0]
        print(f"  {val}: {rate:.2f}% ({hazardous_count}/{count})")

plt.tight_layout()
plt.savefig(os.path.join(eda_dir, 'categorical_vs_target.png'), dpi=300, bbox_inches='tight')
plt.show()
```

```python

# 5. NUMERICAL VARIABLES DISTRIBUTION
print("\n" + "="*40)
print("5. NUMERICAL FEATURES DISTRIBUTION")
print("="*40)

numerical_cols = ['genergy', 'gpuls', 'gdenergy', 'gdpuls', 'energy', 'maxenergy']

fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.ravel()

for idx, col in enumerate(numerical_cols):
    # Histogram with KDE
    data = df_raw[col]
    
    # Skip if all zeros
    if data.sum() == 0:
        axes[idx].text(0.5, 0.5, f'{col}\nAll values are 0', 
                      ha='center', va='center', transform=axes[idx].transAxes)
        axes[idx].set_title(f'{col} (All zeros)')
        continue
    
    # Log transform for better visualization if data is highly skewed
    if data.skew() > 2:
        plot_data = np.log1p(data[data > 0])
        axes[idx].hist(plot_data, bins=50, alpha=0.7, edgecolor='black')
        axes[idx].set_xlabel(f'log1p({col})')
    else:
        axes[idx].hist(data, bins=50, alpha=0.7, edgecolor='black')
        axes[idx].set_xlabel(col)
    
    axes[idx].set_title(f'{col} Distribution (skew={data.skew():.2f})', fontweight='bold')
    axes[idx].set_ylabel('Frequency')
    
    # Print statistics
    print(f"\n{col} statistics:")
    print(f"  Mean: {data.mean():.2f}")
    print(f"  Median: {data.median():.2f}")
    print(f"  Std: {data.std():.2f}")
    print(f"  Min: {data.min():.2f}")
    print(f"  Max: {data.max():.2f}")
    print(f"  Zeros: {(data == 0).sum()} ({(data == 0).mean()*100:.1f}%)")
    print(f"  Skewness: {data.skew():.2f}")

plt.tight_layout()
plt.savefig(os.path.join(eda_dir, 'numerical_distributions.png'), dpi=300, bbox_inches='tight')
plt.show()

# 6. NUMERICAL VARIABLES vs TARGET (Box Plots)
print("\n" + "="*40)
print("6. NUMERICAL FEATURES vs TARGET")
print("="*40)

fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.ravel()

for idx, col in enumerate(numerical_cols):
    # Box plot
    df_box = df_raw[df_raw[col] > 0] if col in ['energy', 'maxenergy'] else df_raw
    
    bp = axes[idx].boxplot([df_box[df_box['class'] == 0][col], 
                            df_box[df_box['class'] == 1][col]],
                           labels=['Safe', 'Hazardous'], patch_artist=True)
    
    # Color boxes
    bp['boxes'][0].set_facecolor('#2ecc71')
    bp['boxes'][1].set_facecolor('#e74c3c')
    
    axes[idx].set_title(f'{col} by Class', fontweight='bold')
    axes[idx].set_ylabel(col)
    
    # Statistical test
    safe_vals = df_box[df_box['class'] == 0][col]
    hazard_vals = df_box[df_box['class'] == 1][col]
    
    if len(hazard_vals) > 0 and len(safe_vals) > 0:
        stat, p_value = stats.mannwhitneyu(safe_vals, hazard_vals, alternative='two-sided')
        print(f"\n{col} - Mann-Whitney U test:")
        print(f"  Safe median: {safe_vals.median():.2f}")
        print(f"  Hazardous median: {hazard_vals.median():.2f}")
        print(f"  p-value: {p_value:.2e} {'(Significant)' if p_value < 0.05 else '(Not significant)'}")

plt.tight_layout()
plt.savefig(os.path.join(eda_dir, 'numerical_vs_target.png'), dpi=300, bbox_inches='tight')
plt.show()
```

```python

# 7. CORRELATION ANALYSIS
print("\n" + "="*40)
print("7. CORRELATION ANALYSIS")
print("="*40)

# Calculate correlations for numerical columns
numerical_cols_all = df_raw.select_dtypes(include=[np.number]).columns.tolist()
correlation_matrix = df_raw[numerical_cols_all].corr()

# Plot correlation heatmap
plt.figure(figsize=(14, 12))
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
sns.heatmap(correlation_matrix, mask=mask, annot=True, fmt='.2f', 
            cmap='RdBu_r', center=0, square=True, linewidths=0.5,
            cbar_kws={"shrink": 0.8})
plt.title('Feature Correlation Matrix', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.savefig(os.path.join(eda_dir, 'correlation_heatmap.png'), dpi=300, bbox_inches='tight')
plt.show()

# Find top correlations with target
target_corr = correlation_matrix['class'].sort_values(ascending=False)
print("\nTop correlations with target (class):")
for feature, corr in target_corr.head(10).items():
    if feature != 'class':
        print(f"  {feature}: {corr:.3f}")
```

```python

# 8. SEISMIC ENERGY ANALYSIS (Gutenberg-Richter Law)
print("\n" + "="*40)
print("8. SEISMIC ENERGY DISTRIBUTION (Gutenberg-Richter)")
print("="*40)

# Filter non-zero energies
energy_data = df_raw[df_raw['energy'] > 0]['energy'].copy()
log_energy = np.log10(energy_data)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Log-energy distribution
axes[0].hist(log_energy, bins=50, alpha=0.7, edgecolor='black', density=True)
sns.kdeplot(log_energy, ax=axes[0], color='red', linewidth=2)
axes[0].set_xlabel('log10(Energy)')
axes[0].set_ylabel('Density')
axes[0].set_title('Energy Distribution (Log Scale)', fontweight='bold')

# Cumulative distribution (Gutenberg-Richter)
sorted_energy = np.sort(energy_data)[::-1]
ranks = np.arange(1, len(sorted_energy) + 1)
axes[1].loglog(sorted_energy, ranks, 'o', alpha=0.5, markersize=3)
axes[1].set_xlabel('Energy')
axes[1].set_ylabel('Cumulative Frequency')
axes[1].set_title('Gutenberg-Richter Relation', fontweight='bold')
axes[1].grid(True, alpha=0.3)

# Fit line for b-value estimation
logE = np.log10(sorted_energy)
logN = np.log10(ranks)
slope, intercept, r_value, p_value, std_err = stats.linregress(logE[:100], logN[:100])
axes[1].plot(sorted_energy[:100], 10**(intercept + slope * logE[:100]), 
            'r--', linewidth=2, label=f'b-value = {-slope:.2f}')
axes[1].legend()

print(f"\nGutenberg-Richter b-value: {-slope:.3f}")
print(f"R² value: {r_value**2:.3f}")

plt.tight_layout()
plt.savefig(os.path.join(eda_dir, 'energy_distribution.png'), dpi=300, bbox_inches='tight')
plt.show()
```

```python

# 9. SHIFT TYPE ANALYSIS (Wet vs Dry)
print("\n" + "="*40)
print("9. SHIFT TYPE ANALYSIS (Wet vs Dry)")
print("="*40)

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Hazard rate by shift
shift_hazard = df_raw.groupby('shift')['class'].mean() * 100
axes[0, 0].bar(shift_hazard.index, shift_hazard.values, color=['#3498db', '#e67e22'])
axes[0, 0].set_title('Hazard Rate by Shift Type', fontweight='bold')
axes[0, 0].set_ylabel('Hazard Rate (%)')
for i, v in enumerate(shift_hazard.values):
    axes[0, 0].text(i, v + 0.5, f'{v:.1f}%', ha='center')

# Energy distribution by shift
for shift_type in ['W', 'N']:
    data = df_raw[df_raw['shift'] == shift_type]['energy']
    data = data[data > 0]
    axes[0, 1].hist(np.log10(data), bins=30, alpha=0.5, label=f'Shift {shift_type}', density=True)
axes[0, 1].set_xlabel('log10(Energy)')
axes[0, 1].set_ylabel('Density')
axes[0, 1].set_title('Energy Distribution by Shift Type', fontweight='bold')
axes[0, 1].legend()

# Bumps by shift
shift_bumps = df_raw.groupby('shift')[['nbumps', 'nbumps2', 'nbumps3', 'nbumps4']].mean()
shift_bumps.plot(kind='bar', ax=axes[1, 0])
axes[1, 0].set_title('Average Number of Bumps by Shift Type', fontweight='bold')
axes[1, 0].set_ylabel('Average Count')
axes[1, 0].legend(title='Energy Class')

# Time series (using index as proxy for time)
axes[1, 1].plot(df_raw.index, df_raw['energy'].rolling(window=100).mean(), 
                label='Energy (100-sample MA)', alpha=0.7)
axes[1, 1].scatter(df_raw[df_raw['class'] == 1].index, 
                   df_raw[df_raw['class'] == 1]['energy'],
                   color='red', s=20, alpha=0.5, label='Hazardous Events')
axes[1, 1].set_xlabel('Sample Index (Time Proxy)')
axes[1, 1].set_ylabel('Energy')
axes[1, 1].set_title('Energy and Hazardous Events Over Time', fontweight='bold')
axes[1, 1].legend()

print(f"\nShift statistics:")
for shift in ['W', 'N']:
    shift_data = df_raw[df_raw['shift'] == shift]
    print(f"\nShift {shift}:")
    print(f"  Samples: {len(shift_data)}")
    print(f"  Hazard rate: {shift_data['class'].mean()*100:.2f}%")
    print(f"  Avg energy: {shift_data['energy'].mean():.2f}")
    print(f"  Avg bumps: {shift_data['nbumps'].mean():.2f}")

plt.tight_layout()
plt.savefig(os.path.join(eda_dir, 'shift_analysis.png'), dpi=300, bbox_inches='tight')
plt.show()
```

```python

# 10. FEATURE ENGINEERING VALIDATION
print("\n" + "="*40)
print("10. DERIVED FEATURES VALIDATION")
print("="*40)

# Check the features we created in processing
derived_features = ['energy_ratio', 'log_energy', 'shift_risk']
occurrence_features = [col for col in df_processed.columns if col.endswith('_occurred')]

print("\nDerived features created:")
for feat in derived_features:
    if feat in df_processed.columns:
        print(f"  ✓ {feat}: mean={df_processed[feat].mean():.3f}, std={df_processed[feat].std():.3f}")

print(f"\nOccurrence indicators created: {len(occurrence_features)}")
print("  " + ", ".join(occurrence_features[:5]) + "...")

# Correlation of derived features with target
if 'energy_ratio' in df_processed.columns:
    print(f"\nCorrelation of derived features with target:")
    for feat in derived_features:
        if feat in df_processed.columns:
            corr = df_processed[feat].corr(df_processed['class'])
            print(f"  {feat}: {corr:.3f}")
```

```python

# 11. SUMMARY STATISTICS REPORT
print("\n" + "="*40)
print("11. EDA SUMMARY REPORT")
print("="*40)

summary_stats = {
    'Total Samples': len(df_raw),
    'Total Features': len(df_raw.columns),
    'Hazardous Events': class_counts[1],
    'Safe Events': class_counts[0],
    'Hazard Rate (%)': df_raw['class'].mean() * 100,
    'Zero Energy Events': (df_raw['energy'] == 0).sum(),
    'Zero Bumps Events': (df_raw['nbumps'] == 0).sum(),
    'Mean Energy (all)': df_raw['energy'].mean(),
    'Mean Energy (hazardous)': df_raw[df_raw['class'] == 1]['energy'].mean(),
    'Max Energy': df_raw['energy'].max(),
    'Wet Shifts (%)': (df_raw['shift'] == 'W').mean() * 100,
    'Dry Shifts (%)': (df_raw['shift'] == 'N').mean() * 100,
}

print("\nKey Statistics:")
for key, value in summary_stats.items():
    if isinstance(value, float):
        print(f"  {key}: {value:.2f}")
    else:
        print(f"  {key}: {value}")

# Save summary to file
summary_df = pd.DataFrame([summary_stats])
summary_df.to_csv(os.path.join(eda_dir, 'eda_summary.csv'), index=False)

```
